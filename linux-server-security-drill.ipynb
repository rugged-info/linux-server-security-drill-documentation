## Documentation

#### Create users for team

```sudo adduser bart
sudo adduser kaleo
sudo groupadd dbs
sudo groupadd webmasters
sudo usermod -a -G dbs bart
sudo usermod -a -G bds kaleo
sudo usermod -a -G webmasters kaleo```

#### Installing security updates

```sudo apt-get upgrade``` 

#### Install ssh

```sudo apt-get install openssh-server```

Configure ssh access

```sudo nano /etc/ssh/sshd.config```

#### Ensure passwords can't be used for remote login

```Change PasswordAuthentication no```

#### Disable root login for ssh 

```PermitRootLogin no```

#### Verify /etc/ssh/sshd.config says protocol 2

#### Configure mysql

```sudo apt-get install mysql-server```

#### Only users of dbs can interact with mysql

```sudo visudo```

#### adjust /etc/sudoers file to allow dbs group to interact with mysql binary

```
#Cmnd alias specification 
Cmnd_Alias    POWER=/usr/bin/mysql```

#### Members of the dbs group may access my sql
```%dbs ALL=POWER``` 

#### Verify success by logging onto 

```sudo mysql -u root -p```

#### Install nginx

``` sudo apt install nginx```

#### Create wrapper scripts for nginx

```
cd /usr/local/bin
touch nginx_start.sh
touch nginx_stop.sh
touch nginx_restart.sh
sudo chmod 744 nginx_start.sh nginx_stop.sh nginx_restart.sh ```

#### Write the scripts

A. 
```
sudo nano nginx_restart.sh

#!/bin/bash

sudo service nginx restart```

B.
``` 
sudo nano nginx_start.sh

#!/bin/bash

sudo service nginx start ```

C.

```
sudo nano nginx_stop.sh

#!/bin/bash

sudo service nginx stop ```


#### Verify scripts work

```
sudo /usr/local/bin/nginx_stop.sh
sudo service nginx status

sudo /usr/local/bin/nginx_start.sh
sudo service nginx status

sudo /usr/local/bin/nginx_restart.sh
sudo service nginx status

```

#### Create new group and new user, then create a script that allows user to start and stop app armor

```
sudo groupadd apparmor
sudo adduser john
sudo usermod -a -G apparmor john
touch /usr/local/bin/apparmor_starter.sh
touch /usr/local/bin/apparmor_stopper.sh
sudo chmod 744 /usr/local/bin/apparmor_starter.sh /usr/local/bin/apparmor_stopper.sh
sudo nano /usr/local/bin/apparmor_starter.sh
    # write script

    #!/bin/bash
    sudo service apparmor start
    sudo service apparmor status

sudo nano /usr/local/bin/apparmor_stopper.sh

    # write script

    #!/bin/bash
    sudo service apparmor stop
    sudo service apparmor status

sudo visudo 

Cmnd_Alias    APPARMOR=/usr/local/bin/apparmor_stopper.sh, /usr/local/bin/apparmor_starter.sh

%apparmor    ALL=APPARMOR ```
